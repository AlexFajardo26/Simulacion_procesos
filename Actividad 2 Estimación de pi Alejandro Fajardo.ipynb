## Actividad 2 Estimaci√≥n de pi

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calcular_pi(N):
    x = np.random.uniform(-1, 1, N)
    y = np.random.uniform(-1, 1, N)
    
    area_cir = x**2 + y**2 <= 1
    pi_est = (np.sum(area_cir) / N) * 4
    return pi_est, x, y, area_cir
```

## N= 10

```python
N = 10
pi_est, x, y, area_cir = calcular_pi(N)


plt.figure(figsize=(6,6))
circulo = plt.Circle((0, 0), 1, color='b', fill=False)
plt.gca().add_patch(circulo)

plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.gca().set_aspect('equal')

plt.scatter(x[area_cir], y[area_cir], color='r', s=0.5)  # Dentro del c√≠rculo
plt.scatter(x[~area_cir], y[~area_cir], color='b', s=0.5) # Fuera del c√≠rculo

# T√≠tulo con la estimaci√≥n de pi
plt.title(f"N = {N}, Calcular de œÄ ‚âà {pi_est:.6f}")
plt.show()
```
