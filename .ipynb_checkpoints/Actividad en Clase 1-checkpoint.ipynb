```python
import yfinance as yf
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
def get_adj_close(tickets, start=None, end=None):
    return yf.download(tickets, start=start, end=end)['Adj Close']
```

```python
names = ['NFLX', 'AMZN', 'MSFT', 'TSLA', 'WMT']
start = '2022-01-01'
```

```python
closes = get_adj_close(tickets=names,
                        start=start)
closes.plot()
```

```python
closes.mean().values
```

```python
NFLX = pd.DataFrame(closes.NFLX)
NFLX.query('NFLX > NFLX.mean()')
```
